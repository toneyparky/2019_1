## 사용법

사용자는 철도로 들어오는 열차의 번호를 입력한다. 열차는 1번부터 9번까지 있다.
철도 스택을 이용해 출구로 나가는 열차는 1번부터 차례대로 나가야 한다. 이를 위해 열차 스택을 이용한다.
화면에는 다음의 내용을 표시한다.
기지에 열차가 한 대씩 들어오는 상황 (3번 열차가 들어옴 : IN 3)
철도 스택에 열차가 들어가는 상황(스택 번호 명시해야 함) (3번 열차가 1번 스택에 들어감 : PUSH 1, 3)
철도 스택에서 열차가 나가는 상황(1번 스택에서 열차가 나감 : POP 1)
대기 장소의 열차를 출구로 내보내는 상황 (OUT 1:1번 열차를 출구로 내보냄)
#### 우리는 적은 횟수만큼 움직이는 것이 좋지만, 
### 최적이 아니어도 괜찮다.

뭔가 스택에 든 것들을 다 한쪽에서 다른쪽으로 옮기며 체크하는 방법으로 하면 최적은 아니어도 스택을 잘 활용하며 풀 수 있을 듯!!!!!!!

## 필요한 것

메서드 : rearrange (이거 하나로 돌린다.)
    trains에서 요소 하나하나인 train을 들여보낼 IN (메서드로 구현하지 않아도 그냥 rearrange안에서 print로 출력가능)
    stack_1, stack_2를 넣을 PUSH
    stack_1, stack_2에서 train을 뺄 POP
    in에서건 stack_1, stack_2에서건 train이 도착지로 갈 때 사용할 OUT
    

변수 : stack_1, stack_2
    trains    ==     (input)

----------------------------------------------------------------------------

## Flow

하나의 메서드에서 진행
rearrange

1. 인풋(trains)을 받아서 rearrange에 파라미터로 전달한다.

2. 출력할 기차의 번호를 체크할 1~10 까지 i 로 돌아가는 for 문

3. trians에서 하나씩 읽는 for문

4-1. 읽은 train == i 이면 IN출력후 바로 OUT하기

4-2. 읽은 train != i 이면 stack의 상태를 확인

5-1. stack_1 과 stack_2가 모두 비어있으면 stack_1에 넣기

5-2. stack_2만 비어있다면 읽은 train을 stack_2에 넣고 stack_1에 있는 요소들을 POP하며 i와 같은지 비교

5-2-1. i와 같다면 OUT해주기

5-2-2. i와 같지 않다면 stack_1에 PUSH한다.

5-3. stack_1만 비어있다면 읽은 train을 stack_1에 넣고 stack_2에 있는 요소들을 POP하며 i와 같은지 비교

5-3-1. i와 같다면 OUT해주기

5-3-3. i와 같지 않다면 stack_2에 PUSH한다.

    + 5-2-1과 5-3-1이 발생하면 trains = trains[index + 1 :] 를 이용하여 trains 자체의 길이를 줄여서 3.의 for문에 다시 들어갈 때 처음부터 train들을 넣지 않도록 해준다.
6. 위의 방식으로 진행하면 trians의 길이가 0이되는 문제가 생기는데 이는  3.의 for문을 else로 분기를하고 앞에  if int(len(trains)) == 0: (15줄)을 하여 더이상 trains에서 train을 읽어서 확인하지 않고 스택에 있는 값들만 한쪽에서 다른 쪽으로 옮기면서 i와 비교하는 방법을 사용한다.












```python
r = rearrange()
r.rearrangeTrain(859347261)
```

```python
''' Train Rearranging Program
'''
class rearrange:
    def __init__(self):
        #stack 초기화
        self.stack_1 = myStack()
        self.stack_2 = myStack()
   
    def rearrangeTrain(self, trains):
        trains = str(trains)
        trains_copy = list(trains)[:]
        count = 0
        for i in range(1, 10): 
            #아래의 for문에서 trains의 길이가 계속 줄어드는데 0이 되면 trains를 더이상 읽지 않고 스택에 있는 것들만 읽기.
            if int(len(trains)) == 0:
                #stack_1과 stack_2가 모두 비었을 때 -> 종료
                if self.stack_1.isEmpty() and self.stack_2.isEmpty():
                    break
                #stack_2만 비었을 때
                elif (not self.stack_1.isEmpty()) and self.stack_2.isEmpty():
                    #train이 i와 같지 않고 stack_2만 비었다면 stack_2에 넣는다.
                    for j in range(0, len(self.stack_1.show())):
                        temp_train = self.stack_1.pop()
                        count += 1
                        print(str(count) + " : POP(" + temp_train + ")")
                        if int(temp_train) == int(i):  #int로 바꾸어져야 비교가 가능하다
                            temp_out = temp_train #mark
                            count += 1
                            print(str(count) + " : OUT(" + temp_train + ")")
                        else:
                            count += 1
                            print(str(count) + " : PUSH(2, " + temp_train + ")")
                            self.stack_2.push(temp_train)
                    if int(temp_out) == int(i):
                        continue
                #stack_1만 비었을 때
                elif self.stack_1.isEmpty() and (not self.stack_2.isEmpty()):
                    #train이 i와 같지 않고 stack_1만 비었다면 stack_1에 넣는다.
                    for j in range(0, len(self.stack_2.show())):
                        temp_train = self.stack_2.pop()
                        count += 1
                        print(str(count) + " : POP(" + temp_train + ")")
                        if int(temp_train) == int(i):  #int로 바꾸어져야 비교가 가능하다
                            temp_out = temp_train #mark
                            count += 1
                            print(str(count) + " : OUT(" + temp_train + ")")
                        else:
                            count += 1
                            print(str(count) + " : PUSH(1, " + temp_train + ")")
                            self.stack_1.push(temp_train)
                    if int(temp_out) == int(i):
                        continue
            #trains를 계속 읽으며 확인하는 과정
            else:   
                for index, train in enumerate(trains):
                    count += 1
                    print(str(count) + " : IN(" + train + ")")
                    """
                    print("i: ", i)
                    print("train: ", train)"""
                    #IN train이 바로 OUT해야 할 경우
                    if int(i) == int(train): #int로 바꾸어져야 비교가 가능하다
                        temp_out = train #mark
                        count += 1
                        print(str(count) + " : OUT(" + train + ")")
                        trains = trains[index + 1 :] #다시 돌아가게끔 시작지점 설정! #mark
                        break
                    #IN train이 stack에 들어가야할 경우
                    else:
                        temp_out = train #mark
                        #stack_1과 stack_2가 모두 비었을 때
                        if self.stack_1.isEmpty() and self.stack_2.isEmpty():
                            count += 1
                            print(str(count) + " : PUSH(1, " + train + ")")
                            self.stack_1.push(train)
                            continue
                        #stack_2만 비었을 때
                        elif (not self.stack_1.isEmpty()) and self.stack_2.isEmpty():
                            #train이 i와 같지 않고 stack_2만 비었다면 stack_2에 넣는다.
                            count += 1
                            print(str(count) + " : PUSH(2, " + train + ")")        
                            self.stack_2.push(train)
                            for j in range(0, len(self.stack_1.show())):
                                temp_train = self.stack_1.pop()
                                count += 1
                                print(str(count) + " : POP(" + temp_train + ")")
                                if int(temp_train) == int(i):  #int로 바꾸어져야 비교가 가능하다
                                    temp_out = temp_train #mark
                                    count += 1
                                    print(str(count) + " : OUT(" + temp_train + ")")
                                else:
                                    count += 1
                                    print(str(count) + " : PUSH(2, " + temp_train + ")")
                                    self.stack_2.push(temp_train)
                            if int(temp_out) == int(i):
                                trains = trains[index + 1 :]
                                break
                        #stack_1만 비었을 때
                        elif self.stack_1.isEmpty() and (not self.stack_2.isEmpty()):
                            #train이 i와 같지 않고 stack_1만 비었다면 stack_1에 넣는다.
                            count += 1
                            print(str(count) + " : PUSH(1, " + train + ")")        
                            self.stack_1.push(train)
                            for j in range(0, len(self.stack_2.show())):
                                temp_train = self.stack_2.pop()
                                count += 1
                                print(str(count) + " : POP(" + temp_train + ")")
                                if int(temp_train) == int(i):  #int로 바꾸어져야 비교가 가능하다
                                    temp_out = temp_train #mark
                                    count += 1
                                    print(str(count) + " : OUT(" + temp_train + ")")
                                else:
                                    count += 1
                                    print(str(count) + " : PUSH(1, " + temp_train + ")")
                                    self.stack_1.push(temp_train)
                            if int(temp_out) == int(i):
                                trains = trains[index + 1 :]
                                break        
```

# Set MyStack Class : 
## method : create, push, pop, peek, is_empty, show, length
Using same class I made for a first assignment.

```python
class myStack:
    def __init__(self):         #create an empty stack (list)
        self.stack = []

    def push(self, item):       #push an item to the stack
        self.stack.append(item)
    
    def pop(self):              #pop an item from the stack
        try:
            return self.stack.pop()
        except IndexError:
            print("Stack is empty")
        
    def isEmpty(self):          #check the stack is empty or not
        return len(self.stack) == 0 #!! no need to make "isFull" (there is no ending point in the stack)
        
    def peek(self):             #get an item from the stack without delete it
        if len(str(self.stack)) == 0:
            raise Exception("Stack empty!")
        else:
            return self.stack[-1]  
    
    def show(self):             #show all items from the stack
        return self.stack
        
```
