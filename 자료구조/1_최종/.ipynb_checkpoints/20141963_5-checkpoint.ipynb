## selectionsort by name

```python
def selection_sort_name(input_array):
    for i in range(len(input_array)-1):
        min_index = i
        #finding a minimum array
        for j in range(i+1, len(input_array)):
            if input_array[j][2] < input_array[min_index][2]:
                min_index = j
        #exchange the minimum array 
        temp_array = input_array[i]
        input_array[i] = input_array[min_index]
        input_array[min_index] = temp_array
    return input_array


```

## selectionsort by student number

```python
def selection_sort_stdnum(input_array):
    for i in range(len(input_array)-1):
        min_index = i
        #finding a minimum array
        for j in range(i+1, len(input_array)):
            if input_array[j][1] < input_array[min_index][1]:
                min_index = j
        #exchange the minimum array 
        temp_array = input_array[i]
        input_array[i] = input_array[min_index]
        input_array[min_index] = temp_array
    return input_array


```

## quicksort by name

```python
def quick_sort_name(input_array):
    input_array_length = len(input_array)
    big_array = []
    small_array = []
    if( input_array_length <= 1):
        return input_array
    else:
        pivot = input_array[0][2]
        #print(pivot)
        for index, array in enumerate(input_array):
            if index == 0:
                pass
            elif array[2] > pivot:
                big_array.append(array)
                #print("big_array: ", big_array)
            elif array[2] <= pivot:
                small_array.append(array)
                #print("small_array: ", small_array)

        '''print("small_array: ", small_array)
        print("pivot: ", pivot)
        print("big_array: ", big_array)'''
        return quick_sort_name(small_array) + [input_array[0]] + quick_sort_name(big_array)
```

## quicksort by student number

```python
def quick_sort_stdnum(input_array):
    input_array_length = len(input_array)
    big_array = []
    small_array = []
    if( input_array_length <= 1):
        return input_array
    else:
        pivot = int(input_array[0][1])
        #print(pivot)
        for index, array in enumerate(input_array):
            if index == 0:
                pass
            elif int(array[1]) > pivot:
                big_array.append(array)
                #print("big_array: ", big_array)
            elif int(array[1]) <= pivot:
                small_array.append(array)
                #print("small_array: ", small_array)
        
        '''print("small_array: ", small_array)
        print("pivot: ", pivot)
        print("big_array: ", big_array)'''
        return quick_sort_stdnum(small_array) + [input_array[0]] + quick_sort_stdnum(big_array)
```

## heapsort by name

```python
def heap_sort_name(input_array): 
    length = len(input_array)
    #print("len(input_array)", len(input_array))
    #print(input_array)
    #make an initial heal
    for i in range(length, -1, -1): 
        adjust_name(input_array, length, i) 
        #print("pass")
  
    #exchange top and tail
    for i in range(length-1, 0, -1): 
        #print(input_array)
        #swap 
        input_array[i], input_array[0] = input_array[0], input_array[i]   
        adjust_name(input_array, i, 0) 
    
    return input_array
```

```python
def adjust_name(input_array, n, i): 
    #set root
    #print("go: ", input_array)
    largest = i   
    left = 2 * i + 1     # left = 2*i + 1 
    right = 2 * i + 2     # right = 2*i + 2 
    '''
    print("i", i)
    print("left", left)
    print("right", right)
    print("largest", largest)
    print(input_array)
    print("~~~~~")'''
    
    if left < n and (input_array[i])[2] < (input_array[left])[2]: 
        largest = left 
  
    #check there is right child of root exists && is greater than root 
    if right < n and input_array[largest][2] < input_array[right][2]: 
        largest = right 
  
    #change root, when it's needed 
    if largest != i: 
        temp = input_array[i]
        input_array[i] = input_array[largest] 
        input_array[largest] = temp  #swap 
  
        #adjust the root. 
        adjust_name(input_array, n, largest) 
```

## heapsort by student name

```python
def heap_sort_stdnum(input_array): 
    length = len(input_array)
    #print("len(input_array)", len(input_array))
    #print(input_array)
    #make an initial heal
    for i in range(length, -1, -1): 
        adjust_stdnum(input_array, length, i) 
        #print("pass")
  
    #exchange top and tail
    for i in range(length-1, 0, -1): 
        #print(input_array)
        #swap 
        input_array[i], input_array[0] = input_array[0], input_array[i]   
        adjust_stdnum(input_array, i, 0) 
    
    return input_array
```

```python
def adjust_stdnum(input_array, n, i): 
    #set root
    #print("go: ", input_array)
    largest = i   
    left = 2 * i + 1     # left = 2*i + 1 
    right = 2 * i + 2     # right = 2*i + 2 
    '''
    print("i", i)
    print("left", left)
    print("right", right)
    print("largest", largest)
    print(input_array)
    print("~~~~~")'''
    
    if left < n and int((input_array[i])[1]) < int((input_array[left])[1]): 
        largest = left 
  
    #check there is right child of root exists && is greater than root 
    if right < n and input_array[largest][1] < input_array[right][1]: 
        largest = right 
  
    #change root, when it's needed 
    if largest != i: 
        temp = input_array[i]
        input_array[i] = input_array[largest] 
        input_array[largest] = temp  #swap 
  
        #adjust the root. 
        adjust_stdnum(input_array, n, largest) 
```
